```python
'''
DATA MANIPULATION FUNCTIONS
'''

# Function to calculate the mean
def calculate_mean(data):
    return sum(data) / len(data)

# Function to calculate the range
def calculate_range(data):    
    return max(data) - min(data)

# Function to calculate variance
def calculate_variance(data):
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - calculate_mean(data))**2
        squared_diff.append( squared_diff_per_x )      
    
    return sum(squared_diff) / len(data)

# Function to calculate standard deviation
def calculate_standard_deviation(data):
    return calculate_variance(data)**0.5

'''
DATASET INITIALIZATION (PRACTICE 1)
'''
print("\nDATA FOR PRACTICE 1\n")

# Initialize dataset, sorting, and data size
data1 = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data   : {data1}")
s_data1 = sorted(data1)
print(f"Sorted : {s_data1}")
size1 = len(data1)
print(f"Length : {size1}\n")

# Calculate all the important values
mean1 = calculate_mean(data1)
print(f"Mean               : {mean1}")
range1 = calculate_range(data1)
print(f"Range              : {range1}")
var1 = calculate_variance(data1)
print(f"Variance           : {var1}")
sdev1 = calculate_standard_deviation(data1)
print(f"Standard Deviation : {sdev1}\n")

'''
HISTOGRAM CREATION (PRACTICE 1)
'''

import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data1, bins=10, edgecolor='black', alpha=0.5, color='green')

# Add a horizontal line
plt.axhline(y=1.5, xmin=4/70*1, xmax=66/70*1, color='purple', linestyle='dashed', linewidth=1, label=f'Range : {range1}')

# Add vertical line
plt.axvline(x=min(data1), color='blue', linestyle='dashed', linewidth=1, label=f'Min : {min(data1):.2f}')
plt.axvline(x=max(data1), color='blue', linestyle='dashed', linewidth=1, label=f'Max : {max(data1):.2f}')
plt.axvline(x=mean1, color='red', linestyle='dashed', linewidth=1, label=f'Mean : {mean1}')

# Add legend
plt.legend()

# Add vertical lines for standard deviation calculation
plt.axvline(x=mean1 - sdev1, color='yellow', linestyle='dashed', linewidth=1, label=f'{mean1-sdev1:.2f}')
plt.axvline(x=mean1 + sdev1, color='yellow', linestyle='dashed', linewidth=1, label=f'{mean1+sdev1:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('GOALS')

# Show the plot
plt.show()

'''
DATASET INITIALIZATION (PRACTICE 2)
'''
print("\nDATA FOR PRACTICE 2\n")

# Initialize dataset, sorting, and data size
data2 = [180, 185, 175, 190, 170, 195, 165]
print(f"Data   : {data2}")
s_data2 = sorted(data2)
print(f"Sorted : {s_data2}")
size2 = len(data2)
print(f"Length : {size2}\n")

# Calculate all the important values
mean2 = calculate_mean(data2)
print(f"Mean               : {mean2}")
range2 = calculate_range(data2)
print(f"Range              : {range2}")
var2 = calculate_variance(data1)
print(f"Variance           : {var2}")
sdev2 = calculate_standard_deviation(data2)
print(f"Standard Deviation : {sdev2}\n")

'''
HISTOGRAM CREATION (PRACTICE 2)
'''

import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data2, bins=10, edgecolor='black', alpha=0.5, color='green')

# Add a horizontal line
plt.axhline(y=1.2, xmin=4/70*1, xmax=66/70*1, color='purple', linestyle='dashed', linewidth=1, label=f'Range : {range2}')

# Add vertical line
plt.axvline(x=min(data2), color='blue', linestyle='dashed', linewidth=1, label=f'Min : {min(data2):.2f}')
plt.axvline(x=max(data2), color='blue', linestyle='dashed', linewidth=1, label=f'Max : {max(data2):.2f}')
plt.axvline(x=mean2, color='red', linestyle='dashed', linewidth=1, label=f'Mean : {mean2}')

# Add legend
plt.legend()

# Add vertical lines for standard deviation calculation
plt.axvline(x=mean2 - sdev2, color='yellow', linestyle='dashed', linewidth=1, label=f'{mean2-sdev2:.2f}')
plt.axvline(x=mean2 + sdev2, color='yellow', linestyle='dashed', linewidth=1, label=f'{mean2+sdev2:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('HEIGHT')

# Show the plot
plt.show()
```
