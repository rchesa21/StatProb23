```python
# PRACTICE 1

'''
DATA MANIPULATION
'''

# Initialize dataset, sorting, and data size
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data:   {data}")
s_data = sorted(data)
print(f"Sorted: {s_data}")
size = len(data)
print(f"Length: {size}\n")

def mean(data): # Find the mean 
    data_mean = sum(data) / len(data)
    return data_mean

def median(data): # Find the median
    mid = len(data)//2
    s_data = sorted(data)

    if len(data) % 2 == 0:
        data_median = (s_data[mid] + s_data[mid+1])/2
    else:
        data_median = s_data[mid]
        
    return data_median

def modus(data): # Find the modus
    from collections import Counter
    counter = Counter(data)
    count_data = dict(counter)

    max_counter = max(list(counter.values()))
    
    data_mode = []
    for k, v in count_data.items():
        if v == max(list(counter.values())):
            data_mode.append(k)

    return data_mode


''' 
CENTRAL TENDENCY OUTPUT
'''

print(f"Mean:   {mean(data):.2f}")
print(f"Median: {median(data)}")
print(f"Mode:   {modus(data)}") 


'''
HISTOGRAM CREATION
'''

import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=100, edgecolor='black', alpha=0.5, color='green')

# Add vertical lines
plt.axvline(x=mean(data), color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean(data):.2f}')
plt.axvline(x=median(data), color='blue', linestyle='dashed', linewidth=1, label=f'Median = {median(data):.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('HEIGHT')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Dalam latihan modul 1.1 diberikan data tinggi badan siswa kelas A yang berjumlah 11 siswa. Diketahui dalam data terdapat satu outlier yang mempengaruhi mean, serta jumlah modus yang terlalu banyak karena samanya jumlah frekuensi. Sehingga, dapat diambil kesimpulan bahwa pemusatan yang paling cocok digunakan adalah MEDIAN.

```python
# PRACTICE 2

'''
DATA MANIPULATION
'''

# Initialize dataset, sorting, and data size
data = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]
print(f"Data:   {data}")
s_data = sorted(data)
print(f"Sorted: {s_data}")
size = len(data)
print(f"Length: {size}\n")

def mean(data): # Find the mean 
    data_mean = sum(data) / len(data)
    return data_mean

def median(data): # Find the median
    mid = len(data)//2
    s_data = sorted(data)

    if len(data) % 2 == 0:
        data_median = (s_data[mid] + s_data[mid+1])/2
    else:
        data_median = s_data[mid]
        
    return data_median

def modus(data): # Find the modus
    from collections import Counter
    counter = Counter(data)
    count_data = dict(counter)

    max_counter = max(list(counter.values()))
    
    data_mode = []
    for k, v in count_data.items():
        if v == max(list(counter.values())):
            data_mode.append(k)

    return data_mode


''' 
CENTRAL TENDENCY OUTPUT
'''

print(f"Mean:   {mean(data):.2f}")
print(f"Median: {median(data)}")
print(f"Mode:   {modus(data)}") 


'''
HISTOGRAM CREATION
'''

import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=20, edgecolor='black', alpha=0.5, color='green')

# Add vertical lines for mean and median
plt.axvline(x=mean(data), color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean(data):.2f}')
plt.axvline(x=median(data), color='blue', linestyle='dashed', linewidth=1, label=f'Median = {median(data):.2f}')

# Add legend
plt.legend()

# Add vertical lines for all modes
mode_values = modus(data)
for i in range(0, len(mode_values)):
    plt.axvline(x=mode_values[i], color='yellow', linestyle='dashed', linewidth=1, label=f'Mode {i+1}= {mode_values[i]:.2f}')    

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('SCORE')

# Show the plot
plt.show()
```
